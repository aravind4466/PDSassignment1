```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
dataset=pd.read_csv('/content/clean_data/clean_data1.csv')
```

```python
#Visualisation1
plt.bar(dataset['gender'].value_counts().index,
       dataset['gender'].value_counts().values,
       color = ['green', 'blue'])
plt.title('Gender Distribution')
plt.xticks(dataset['gender'].value_counts().index,['Female','Male'])
```

```python
#visualization 2
parent_education_levels = dataset["parental level of education"].unique()
parent_edu_order = ["master's degree","bachelor's degree",  "associate's degree",'some college',  'high school', 'some high school']
df_filtered = dataset[dataset["parental level of education"].isin(parent_edu_order)]
counts = df_filtered.groupby(["parental level of education", "race/ethnicity"]).size().unstack(fill_value=0)
counts.plot(kind="bar", stacked=True, figsize=(10, 6))
```

```python
#visualization 3:
figure, axis = plt.subplots(2,2)
sns.histplot(dataset['math score'],ax=axis[0,0],label='math score')
sns.histplot(dataset['reading score'],ax=axis[0,1],label='reading score')
sns.histplot(dataset['writing score'],ax=axis[1,0],label='writing score')
sns.histplot(dataset['Total score'],ax=axis[1,1],label='Total score')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Frequency distribution of scores')
```

```python

# Data visualization 4
df_masters = dataset[dataset['parental level of education'] == "master's degree"]
race_counts = df_masters['race/ethnicity'].value_counts()
# Create a pie chart
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Race/Ethnicity Among Students Whose Parents Have a Masters Degree")
plt.show()
```
